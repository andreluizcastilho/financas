## Aula 7.1. Armazenamento de dados de mercado

### Exemplo 1 - Armazenamento de dados coletados da API Yahoo! Finance

Importar a biblioteca yfinance

```python
!pip install yfinance
import yfinance as yf
```

Definimos o instrumento a ser requisitado

Para isso, primeiro conferir o ticker do instrumento acessando https://finance.yahoo.com/

```python
MBLY3 = yf.Ticker("MBLY3.SA")
```

```python
df_hist = MBLY3.history(period="5y", interval="1d")
```

```python
df_hist.head()
```

```python
df_hist.tail()
```

```python
df_hist = df_hist[
    [
        "Open",
        "High",
        "Low",
        "Close",
        "Volume"
    ]    
]
```

```python
df_hist
```

```python
df_hist.reset_index(inplace=True)
```

```python
df_hist
```

```python
df_hist.columns = [
    "data_ref",
    "preco_abertura",
    "preco_maxima",
    "preco_minima",
    "preco_fechamento",
    "volume"
]
```

```python
df_hist
```

```python
df_hist["ativo"] = "MBLY3.SA"
```

```python
df_hist
```

```python
# PODE SER QUE SEJA NECESSÁRIO INSTALAR ESSAS BIBLIOTECAS NO AMBIENTE VIRTUAL XP_CRV PELO CMD
!pip install pymysql
!pip install SQLAlchemy 
from sqlalchemy import create_engine
```

```python
pwd = ''
con = create_engine(
    f'mysql+pymysql://root:{pwd}@localhost:3306/dados_mercado'
)
```

```python
df_hist.to_sql(
    name="precos_acoes",
    con=con,
    if_exists='replace'
)
```
